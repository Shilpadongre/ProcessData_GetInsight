```python
#preprocessing data
def preprocess_data(df):
  #split combined data in the seperate column
  df[['district_code','state','district_name']]=df['State & District'].str.split(',',2,expand=True)
  df.drop(columns=['State & District'],inplace=True)

  return df

#Preprocess the data
df1=preprocess_data(df.copy())

#display preprocessed data
print("Preprocessed Data:")
df1.head()
```

```python
df1['state']=df1['state'].str.replace('state_name:',"")
df1['district_code']=df1['district_code'].str.replace('district_code:',"")
df1['district_name']=df1['district_name'].str.replace('district_name:',"")
```
